# Filtering

Filtering dataframe removes unwanted data; we can do this in multiple ways. Here, there are some examples.

For example, let us say we want to seize the data of the Gentoo species of penguins. We set up a filter in the following way. <br>
At first, we make a variable corresponding to the data we want. In this case, the data is about the Gentoo species

```python
filter = (df['species'] == 'Gentoo')
```

... now we apply to it to our dataframe

```python
df[filter]
```

We can also do this using the <b>.loc</b> method

```python
df.loc[filter]
```

When <b>.loc</b> is used, we can become more specific with our filter by adding specific columns

```python
df.loc[filter,['island','sex']]
```

Now, we want data corresponding not only to a specific species but also to the sex of penguins - especially the species Adelie and Male. 

```python
filter = (df['species'] == 'Adelie') & (df['sex'] == 'MALE')
df.loc[filter]
```

Or ... we want a dataframe with the penguins from the Dream or Torgersen islands

```python
filter = (df['island'] == 'Torgersen') | (df['island'] == 'Dream')
```

```python
df[filter]
```

Or ... we want to work with data about penguins with a flipper length greater than 200. So, we could filter out the data like this

```python
filter = df['flipper_length_mm'] > 200
```

```python
df[filter]
```

Now, we would like to work with data about penguins that have a flipper length greater than 200mm AND a body mass less than 4000g  

```python
filter = ((df['flipper_length_mm'] > 200) & (df['body_mass_g'] < 4000))
```

```python
df[filter]
```

Let us look at filtering when working with strings ...

We want to work with data from species that have the letter <b>e</b> in their names; we would use the <b>.str.contains('letter')</b>.

```python
filter = df['species'].str.contains('e')
```

```python
df[filter]
```

Or ... we want to work with data about penguins from an island that contains a string of letters; we would use the <b>.str.contains('string')</b>.

```python
filter = df['island'].str.contains('ea')
```

```python
df[filter]
```

If we want to work with data from a specific column that has a string starting with a specific letter or string;
we would use the <b>.str.startswith('letter')</b> or <b>.str.startswith('string')</b>.

```python
filter = df['species'].str.startswith('A')
```

```python
df[filter]
```

Another approach: if we want to work with data that ends with a string - we use <b>.str.endswith('letter')</b> or <b>.str.endswith('string')</b>. 

For example, we have an array of column names we would like to work with - we could use the <b>.isin(array)</b> method.

```python
array = ['Torgersen','Dream']
```

```python
filter = df['island'].isin(['Torgersen','Dream'])
```

```python
df[filter]
```

# Cleaning Data

We can remove rows with missing values using the <b>dropna()</b> method. The default parameters working in the background are <b>dropna(axis='index', how='any')</b>. We can set the axis to either 'index' (drop the row) or 'columns' (drop the column). The 'any' parameter says that 'any' row or column with a missing value gets dropped. We can change this to 'all,' meaning we only remove a missing value if the entire row or column is filled with missing values. 

```python
df
```

```python
df.dropna(axis='index', how='any')
```

```python
df.dropna(axis = 'index', how = 'all')
```

```python
df.dropna(axis = 'columns', how = 'any')
```

```python
df.dropna(axis = 'columns', how = 'all')
```

You can also pass a subset into the <b>dropna()</b> method. Then only this subset is checked for missing values to drop across a specific column: <b>dropna(axis='index', subset =['sex'])</b>. Note that the square brackets imply a list of specified columns. Also, the 'how' parameter is not needed here because we are only dropping rows that are missing the values in the 'sex' column, which makes the' how' parameter redundant.

```python
df.dropna(axis = 'index', subset = ['sex'])
```

The library <b>numpy</b> is helpful here. This library provides tools to directly refer to the missing 'NaN' values.

```python
import numpy as np
```

Now, we can have missing values written differently due to different software - instead of 'NaN,' it can be written as 'missing' or 'not available.'

It can be fixed using the replace method, for example, <b>df.replace('name-of-the-missing-variable', np.NaN, inplace=True)</b>.

It replaces the missing variable names with 'NaN,' which can then be removed using the <b>dropna()</b> method.

The reverse of this operation is the <b>fillna()</b> method.
We insert, as an argument, a string or integer, which we would like to replace the NaN value with - <b>fillna("NOT PRESENT")</b>.

Another convenient method is the <b>isna()</b>. It returns a dataframe where the values are replaced with true or false values, with true being a NaN value.

Note that the missing value NaN is a float. If we need to perform numerical operations with some data that have missing values, we need to convert them to a float to account for the NaN values. We  convert types by using type casting: <b>df['column-name'] = df['column-name'].astype(float)</b>.

We can also quickly identify the strings within a column using the <b>unique()</b> method for a series. It lets us identify certain strings that cannot be turned into a 'float' - we can replace them with the <b>replace()</b> method.
