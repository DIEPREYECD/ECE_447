```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

Let's get a feel for the dataframe we will be using in this tutorial

```python
df = pd.read_csv('data/penguins_lter.csv')
df.head(5)
```

```python
df.tail(5)
```

```python
df_Sex = df.drop(axis = 0, index = 336)
```

To see more details about each function press the Shift + Tab keys to get on the cell with the function to see more infomation about the function.

# Scatter Plot

```python
sns.scatterplot(data = df_Sex, x = 'Body Mass (g)', y = 'Flipper Length (mm)')
```

scatterplot() plots a scatter plot of the relationship between x and y, x and y being different subsets of a dataframe, there are muliple different parametrs that can be used to get a better guage of your data. An example of one is hue, the hue parameter takes a column name and color enncodes it.

```python
sns.scatterplot(data = df_Sex, x = 'Body Mass (g)', y = 'Flipper Length (mm)', hue ='Sex')
```

# Distribution Plot

```python
sns.displot(df_Sex, x ='Body Mass (g)',hue = 'Sex')
```

A distribution plot is used to shows the distribution of a set number numeric values against an attribute.

Find more details about distribution plots here : https://seaborn.pydata.org/generated/seaborn.distplot.html

# Joint Plot

```python
sns.jointplot(x = 'Body Mass (g)' , y = 'Flipper Length (mm)', data = df)
```

A joint plot plots a scatter plot of x and y whilst also plotting a distribution plot of x and y.

Find more details about joint plots here : https://seaborn.pydata.org/generated/seaborn.jointplot.html

# KDE Plot

```python
sns.kdeplot(df['Culmen Depth (mm)'])
```

Kde(Kernel density estimate) is a plot that is used to visualize the distribution of observation in a data set. The KDE represents data using a continous probability density curve in one or more dimensions.

Find more details about KDE plots here : https://seaborn.pydata.org/generated/seaborn.kdeplot.html

# Pair Plot

```python
sns.pairplot(data = df, hue = 'Species')
```

A pair plot plots all the columns that have numerical values in them against each other.

Find more details about pair plots here : https://seaborn.pydata.org/generated/seaborn.pairplot.html

# Bar Plot

```python
sns.barplot(x = 'Sex', y = 'Body Mass (g)', data = df_Sex)
```

Bar plot is used to show the relationship between a categorical column and a numerical column, it plots the the  estimator of the chosen categorical column. By default the estimator is the mean however you can change it to other estimators for example the median and the varience.

Find more details about Bar plots here : https://seaborn.pydata.org/generated/seaborn.barplot.html

```python
sns.barplot(x = 'Sex', y = 'Body Mass (g)', data = df_Sex, estimator = np.var)
```

# Count Plot

```python
plt.figure(figsize = (10.7,5))
sns.countplot(x = 'Species', data = df)
```

The count plot plots the number of occurences of a categorical column

Find more details about Count plots here : https://seaborn.pydata.org/generated/seaborn.countplot.html

# Box Plot

```python
plt.figure(figsize = (10.7,5))
sns.boxplot(x = 'Species', y = 'Body Mass (g)', data = df_Sex,hue ='Sex')
```

A box plot (or box-and-whisker plot) shows the distribution of quantitative
data in a way that facilitates comparisons between variables or across
levels of a categorical variable. The box shows the quartiles of the
dataset while the whiskers extend to show the rest of the distribution,
except for points that are determined to be "outliers" using a method
that is a function of the inter-quartile range.

Information was gathered from : https://seaborn.pydata.org/generated/seaborn.boxplot.html

It is also possible to manipulate the size of your plots using matplot lib function plt.figure(figure = (x,y)), x being the width of the plot and y being the height of the plot.

Find more details about Box plots here : https://seaborn.pydata.org/generated/seaborn.boxplot.html

# Violin Plot

```python
plt.figure(figsize = (10.7,5))
sns.violinplot(x = 'Species', y = 'Body Mass (g)', data = df_Sex,hue ='Sex')
```

A violin plot plays a similar role as a box and whisker plot. It shows the
distribution of quantitative data across several levels of one (or more)
categorical variables such that those distributions can be compared. Unlike
a box plot, in which all of the plot components correspond to actual
datapoints, the violin plot features a kernel density estimation of the
underlying distribution.

Information was gotten from : https://seaborn.pydata.org/generated/seaborn.violinplot.html

Find more details about violin plots here : https://seaborn.pydata.org/generated/seaborn.violinplot.html

# Strip Plot

```python
plt.figure(figsize = (10.7,5))
sns.stripplot(x = 'Species', y = 'Body Mass (g)', data = df_Sex, hue = 'Sex')
```

A strip plot is essentially  a single axis scatteer plot that visualizes the distribution of a number of individual one dimensional values.

Find more details about strip plots here : https://seaborn.pydata.org/generated/seaborn.stripplot.html

# Swarm Plot

```python
plt.figure(figsize = (10.7,5))
sns.swarmplot(x = 'Species', y = 'Body Mass (g)', data = df_Sex)
#sns.violinplot(x = 'Species', y = 'Body Mass (g)', data = df_Sex)
```

A swarm plot is very similar to a strip plot, however in a swarm plot the points do not obscure each other.

Find more details about swarm plots here : https://seaborn.pydata.org/generated/seaborn.swarmplot.html

Here's a neat trick that will help visualize data better, you can combine two plots together like this:

```python
plt.figure(figsize = (10.7,5))
sns.swarmplot(x = 'Species', y = 'Body Mass (g)', data = df_Sex, color = 'white')
sns.violinplot(x = 'Species', y = 'Body Mass (g)', data = df_Sex)
```

# Heat Map

```python
# correlation = df.corr()
correlation = df.corr(numeric_only=True)
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(correlation, annot=True, cmap = 'Blues')
```

Here we have a heat map that displays the correlation of each numerical column against each other. In statistics correlation measures the extent to which two variables are linearly dependent.
