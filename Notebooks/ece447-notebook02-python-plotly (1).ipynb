```python
figure = px.scatter(df_clean, x = 'body_mass_g', y = 'culmen_length_mm', color = 'sex')
figure.show()
```

Here we have modified the plot to express more things, by setting the 'size' parameter equal to 'flipper_length_mm' the size of each point now scales with flipper length. There is also the 'hover_data' parameter which is used to show extra data about each point, and finally the hover_name parameter is used to emphasize certain data about each point.

```python
figure = px.scatter(df_clean, x = 'body_mass_g', y = 'culmen_depth_mm', color = 'island',
                    size = 'flipper_length_mm', hover_data = ['culmen_depth_mm','sex'],hover_name = 'species')
figure.show()
```

The color parameter may also be used for continuous data rather than just categorical data.

```python
figure = px.scatter(df_clean, x = "body_mass_g", y = "culmen_length_mm", color = 'flipper_length_mm', hover_data=['sex'] )
figure.show()
```

Similar to Seaborn's pairplot function plotly's scatter_matrix() function can be used to plot all the numerical data against each other, however, you can specify which columns to apply using the dimension parameter.

```python
# if pandas >= 2.0
pd.DataFrame.iteritems = pd.DataFrame.items
```

```python
figure = px.scatter_matrix(df_clean, dimensions = ['culmen_length_mm', 'culmen_depth_mm', 'flipper_length_mm', 'body_mass_g'], color = 'species')
figure.update_layout(width = 800, height = 1000)
figure.show()
```

Additionally you can remove plots that plot columns against itself using the 'figure.update_traces(diagonal_visible = False)' function.

```python
figure = px.scatter_matrix(df_clean, dimensions = ['culmen_length_mm', 'culmen_depth_mm', 'flipper_length_mm', 'body_mass_g'], color = 'species')
figure.update_traces(diagonal_visible = False)
figure.update_layout(width = 800, height = 1000)
figure.show()
```

You can generate a three-dimensional scatter plot. This is similar to the two-dimensional version as they have similar parameters except for the extra z parameter.

```python
figure = px.scatter_3d(df_clean, x = 'body_mass_g', y = 'sex', z = 'culmen_length_mm', 
                       color = 'species', size = 'flipper_length_mm',
                       hover_data = ['culmen_depth_mm','sex'],hover_name = 'island')
figure.show()
```

# Histogram / Bar Graph

Using the px function and the bar method we can create a bar graph and specify the axis with 'x' and 'y' parameters.

```python
figure = px.bar(df_clean, x = 'body_mass_g', y = 'flipper_length_mm')
figure.show()
```

We can graph multiple bar graphs by inserting a list of series into the y parameter

```python
figure = px.bar(df_clean, x = 'body_mass_g', y = ['culmen_length_mm', 'culmen_depth_mm', 'flipper_length_mm'], title = 'Wide Example')
figure.show()
```

Here we change the graph by using colours to separate the different islands.

```python
figure = px.bar(df_clean, x='sex', y = 'body_mass_g', color = 'island')
figure.show()
```

Histograms are very similar to bar graphs. For the most part you can swap between them by simply switching the bar method for the histogram method.

```python
figure = px.histogram(df_clean, x = 'island', y = 'flipper_length_mm', color = 'sex', barmode = 'group')
figure.show()
```

```python
figure = px.bar(df_clean, x = 'culmen_length_mm', y = 'body_mass_g', color = 'sex', barmode = 'group', 
               facet_row = 'species', facet_col = 'island')
figure.show()
```

# 2D Histogram

A 2D histogram which is also know as a density heatmap is the 2-D visualization of a histogram, it is useful visualizing the relationship of one attribute aginst another. 

Below, there is a simple example of a 2-D Histogram. The bright spots are where there is a high count of penguins with that specific body mass and flipper length.

```python
figure = px.density_heatmap(df_clean, x = 'body_mass_g', y = 'flipper_length_mm')
figure.show()
```

You can also plot the 2-D histogram with the respective axes distribution plots

```python
figure = px.density_heatmap(df_clean, x = 'flipper_length_mm', y = 'culmen_depth_mm',
                            marginal_x = 'histogram', marginal_y = 'histogram')
figure.show()
```

Or you can add the count for each bin to get a better feel for the 2-D histograms distribution.

```python
figure = px.density_heatmap(df_clean, x = 'culmen_depth_mm', y = 'culmen_length_mm',text_auto = True)
figure.show()
```

# Double Y Axis Plot

Here we will be working with two different scaled y-axis.

```python
# initialize list of lists
data = [[1, 1, 60], [2, 2, 50], [3, 3, 40], [4, 4, 30], [5, 5, 20], [6, 6, 10]]
 
# Create the pandas DataFrame
df_sample = pd.DataFrame(data, columns=['x', 'y1', 'y2'])
df_sample
```

In plotly you can use two differently scaled y-axis to the same x-axis by using the 'make_subplot()' function. Additionally, using the '.add_trace()' method you can assign different line plots to the different y-axis.

```python
figure = make_subplots(specs = [[{'secondary_y' : True}]])

figure.add_trace(
    go.Scatter( x = df_sample['x'], y = df_sample['y1'], name = 'yaxis data 1'),
    secondary_y = False
)

figure.add_trace(
    go.Scatter( x = df_sample['x'], y = df_sample['y2'], name = 'yaxis data 2'),
    secondary_y = True
)

figure.update_layout(title_text = 'Double Y Axis Example')

figure.update_xaxes(title_text = 'x axis title')

figure.update_yaxes(title_text = 'primary y axis', secondary_y = False)
figure.update_yaxes(title_text = 'secondary y axis', secondary_y = True)

figure.show()
```

# Interactivity

Here we will offer some of the more interactive features plotly has to offer.

First we will demonstrate the custom button feature:

## Custom Buttons

The generation of the buttons is accomplished by setting a buttons function to plot body mass (x-axis) vs flipper length (y-axis) in a scatter plot and in a bar plot.

```python
plot = go.Figure(data = [go.Scatter(x = df['body_mass_g'], y = df['flipper_length_mm'], mode ='markers')])

# Making the buttons
my_buttons = list([  
    
    dict(args =['type','scatter'] , label = 'Scatter Plot', method = 'restyle' ),
    dict(args =['type', 'bar'] , label = 'Bar Plot', method = 'restyle' )
    ])
# Your buttons are actually a list of dictionaries that describe what type of button we are dealing with.

plot.update_layout(
    updatemenus = [dict(type = 'buttons', buttons = my_buttons)]
)

plot.show()
```

## Sliders

Using another built-in data set that contains information about life expectancy and GDP per-capita, and using the 'animation_frame' parameter, you can create a slider that iterates through the data.

```python
df_cnt = px.data.gapminder()

px.scatter(df_cnt, x="gdpPercap", y="lifeExp", animation_frame="year",
           size="pop", color="continent", hover_name="country",
           log_x=True, size_max=55, range_x=[100,100000], range_y=[25,90])
```

## Range Slider 

```python
df_stocks = px.data.stocks()
df_stocks
```

```python
px.line(df_stocks, x = 'date', y = 'AAPL', labels={'x':'Date', 'y':'Price'})
```

With a ranged slider you can select and narrow down different ranges of data.

```python
x = df_stocks['date']
y = df_stocks['AAPL']

plot = go.Figure(data = [go.Scatter(x = x, y = y, mode = 'lines')])


plot.update_layout( xaxis = dict(rangeselector = dict(buttons = list([dict(count = 1, step = 'year', stepmode = 'backward')])),
                                 rangeslider = dict(visible = True)
                                )
                  )

plot.show()
```

By creating a data list of dictionaries we can specify a filter and plot specified values

```python
import plotly.io as pio

index = df_clean.index
bodymass = df_clean['body_mass_g']

data = [dict(
  type = 'scatter',
  x = index,
  y = bodymass,
  mode = 'markers',
  transforms = [dict(
    type = 'filter',
    target = 'y',
    operation = '>', #look to see for a range
    value = 4000
  )]
)]

layout = dict(
    title = 'bodymass > 4000'
    
)

fig_dict = dict(data=data, layout=layout)


pio.show(fig_dict, validate=False)
```
